```python
# Ques:7) Mark Sheet And Grade Calculator

eng = int(input(" enter your English Marks"));
isl = int(input(" enter your Islamiyat Marks"));
sci = int(input(" enter your Science Marks"));
math = int(input(" enter your Math Marks"));
comp = int(input("  enter your Computer Marks"));

if eng<=100 and isl<=100 and sci<=100 and math<=100 and comp<=100:
    total = eng+isl+sci+math+comp;
    totalMarks = 500;
    marks = total*100/totalMarks;
    
    if marks <= 100 and marks >= 90:
        print('your total marks is: ',total,' and your grade is A+');
    elif marks <= 90 and marks >= 80:
        print('your total marks is: ',total,' and your grade is A');
    elif marks <= 80 and marks >= 70:
        print('your total marks is: ',total,' and your grade is B');
    elif marks <= 70 and marks >= 60:
        print('your total marks is: ',total,' and your grade is C');
    elif marks <= 60 and marks >= 50:
        print('your total marks is: ',total,' and your grade is D');
    elif marks <= 50 and marks >= 36:
        print('your total marks is: ',total,' and your grade is E');
    elif marks <= 36 and marks >= 0:
        print('your total marks is: ',total,' you are fail please best prectice');
else:
    print("please enter your correct Marks!")
```
